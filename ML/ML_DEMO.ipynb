# 사이킷런을 이용하여 붓꽃(Iris) 품종 예측하기

```python
# 필요한 모듈 로드
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

- 데이터 로딩


```python
# 사이킷런 제공하는 Iris 데이터 로딩
iris = load_iris()

type(iris)
```

```python
iris
```

```python
'''
# feature(특성) : 예측을 수행할 때 사용되는 입력(독립, 설명) 변수
# target(label, 정답) : 예측을 수행할 때 사용되는 출력(종속, 목적, 반응) 변수
------------------------------------------------------------------
# Bunch 객체 데이터
- data : 피처의 데이터 세트
- target : 레이블(정답)의 데이터 세트
- target_names : 레이블의 이름
- feature_names : 피처의 이름
- DESCR : 데이터 세트에 대한 설명

'''
```

```python
# 데이터 세트에서 필요한 정보를 따로 추출해서 확인

# 피처데이터 확인
print(iris.data)
print(iris.feature_names)
```

```python
# target 데이터 확인

print(iris.target)
print(iris.target_names)
```

```python
# 데이터프레임 형태로 변경

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head()
```

```python
# 데이터 기본 요약 정보
iris_df.info()
```

```python
# 학습용 데이터세트와 테스트용데이터 세트를 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state=11)   # random_state 언제나 일정하게 셔플하기 위해 지정해줌
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 모델 생성 - 결정트리알고리즘
from sklearn.tree import DecisionTreeClassifier


# 모델 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)
```

```python
# 모델 학습 - 지도학습
dt_clf.fit(X_train, y_train)
```

```python
# 예측 - 테스트데이터 사용(학습에 사용하지 않은 새로운 데이터)
pred = dt_clf.predict(X_test)

pred
```

```python
# 평가 - 모델이 예측한 결과를 바탕으로 이 모델의 예측 성능을 확인
# - 정확도(Accuracy)
from sklearn.metrics import accuracy_score

print('학습 모델의 예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
```
